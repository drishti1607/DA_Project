```python
dataset = pd.read_csv('Karnataka_Data.csv')
list2 = ['BANGALORE COMMR.']
rural2 =dataset.loc[dataset['DISTRICT'].isin(list2)]
list3 = ['BANGALORE RURAL']
rural3 =dataset.loc[dataset['DISTRICT'].isin(list3)]
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

ax = rural2.plot(x='Year',y='Rape',legend=False)
ay=rural3.plot(x='Year',y='Rape',ax=ax,legend=False)
red_patch = mpatches.Patch(color='red', label='URBAN')
blue_patch = mpatches.Patch(color='blue', label='RURAL')
plt.legend(handles=[blue_patch,red_patch])
plt.title("Rural vs Urban Rape Comparison - Karnataka",bbox={'facecolor':'0.8', 'pad':10},y=-0.3)
plt.show()
```
